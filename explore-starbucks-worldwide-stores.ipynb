Starbucks, American company that is the largest coffeehouse chain in the world. Its headquarters are in Seattle, Washington.

The first Starbucks opened in Seattle, Washington, on March 31, 1971, by three partners who met while they were students at the University of San Francisco: English teacher Jerry Baldwin, history teacher Zev Siegl, and writer Gordon Bowker were inspired to sell high-quality coffee beans and equipment by coffee roasting entrepreneur Alfred Peet after he taught them his style of roasting beans. The company took the name of the chief mate in the book Moby-Dick: Starbuck, after considering “Cargo House” and “Pequod”. Bowker recalls that Terry Heckler, with whom Bowker owned an advertising agency, thought words beginning with “st” were powerful. The founders brainstormed a list of words beginning with “st”. Someone pulled out an old mining map of the Cascade Range and saw a mining town named “Starbo”, which immediately put Bowker in mind of the character “Starbuck”.

Let's visualize NY city on folium with the help of NY city coordinates with Stamen Toner and Stamen terrain background. 
Circle markers are used to point the precise location on the map.Code is also written by using feature group function. 

```python
#folium visualization for Newyork city coordinates.

a =folium.Map(
    location=[48.1667, 100.1667],
    tiles='Stamen Toner',
    zoom_start=6)
a
    
```

```python
# folium CircleMarker -to mark the stores of Newyork co-ordinates.
folium.CircleMarker(
    # radius of CircleMarker
    radius=15,
    # Coordinates of location Newyork 
    location=[40.73,-73.93],
    popup='Newyork',
    # color of CircleMarker
    color='crimson',
    fill=False,
).add_to(a)

#folium circle Marker to map the co-ordinmates of location Manhattan in Newyork.  
folium.CircleMarker(
    # Coordinates of location Manhattan
    location=[-40.78, -73.97],
    # radius of CircleMarker
    radius=15,
    popup='Manhattan',
    # color of CircleMarker
    color='#3186cc',
    fill=True,
    fill_color='#3186cc'
).add_to(a)
a
```

```python
#By using feature group function().
can_map = folium.Map(location=[40.78, -73.97],zoom_start=4,tiles='Stamen Toner')

NY = folium.FeatureGroup()

NY.add_child(folium.features.CircleMarker([40.78, -73.97],radius=5,color='red', fill_color='Red'))

can_map.add_child(NY)

folium.Marker([40.78, -73.97],popup='NY').add_to(can_map)
can_map
```

```python
l = folium.Map(
    location=[35.0167, -102.1667],
    zoom_start=4.5,
    tiles='Stamen Terrain'
)

folium.Marker(
    location=[40.78, -73.977],
    popup=folium.Popup(max_width=450).add_child
).add_to(l)

folium.Marker(
    location=[31.96, -99.91],
    popup=folium.Popup(max_width=450).add_child
).add_to(l)

folium.Marker(
    location=[38.906, -77.03],
    popup=folium.Popup(max_width=450).add_child
).add_to(l)

folium.Marker(
    location=[36.77, -119.41],
    popup=folium.Popup(max_width=450).add_child
).add_to(l)

# Add line to map
folium.PolyLine(locations = [(40.78, -73.97), (31.96,-99.91), (38.90,-77.03), (36.97,-119.41), (35.01,-102.1667)], 
                line_opacity = 0.5).add_to(l) 

l
```

US Statewise Starbucks Store Counts 

Scatter Map ¶
I have first calculated the US states Starbucks store counts.
I have then taken an average of latitude and longitude for the location of stores in each state. 
I am then plotting a scatter plot on it using CircleMarker where circle size represents a number of stores.

```python
starbucks_us = US_data_df[US_data_df["Country"] == "US"].reset_index()
starbucks_us_state_count = starbucks_us.groupby(by="State/Province").count()[["index"]].rename(columns={"index":"Count"}).reset_index()
starbucks_us_lat_lon = starbucks_us.groupby(by="State/Province").mean()[["Latitude", "Longitude"]].reset_index()
starbucks_us = starbucks_us_state_count.merge(starbucks_us_lat_lon, right_on="State/Province", left_on="State/Province").reset_index()

center_lat = starbucks_us.mean().Latitude
center_long = starbucks_us.mean().Longitude

m = folium.Map(location=[center_lat, center_long], zoom_start=4, tiles="Stamen Toner")

for idx in starbucks_us.index:
    row = starbucks_us.iloc[idx]
    lat, long = row.Latitude, row.Longitude
    folium.CircleMarker(
        location=[lat, long],
        radius = float(row.Count/100),
        draggable = False,
        color="red",
        popup = row["State/Province"],
        tooltip = str(row["State/Province"]) + ", " + str(row["Count"]),
        fill_color="yellow"
        ).add_to(m)
m
```

New York City Starbucks Locations Map with Markers ¶


Below are markers on the new york city map where each marker represents the Starbucks store. I have also used popup and tooltip which displays store address.

```python
starbucks_ny = US_data_df[US_data_df["City"] == "New York"].reset_index()

center_lat = starbucks_ny.mean().Latitude
center_long = starbucks_ny.mean().Longitude

m = folium.Map(location=[center_lat, center_long], zoom_start=11.5, )#tiles="Stamen Toner")

for idx in starbucks_ny.index:
    row = starbucks_ny.iloc[idx]
    lat, long = row.Latitude, row.Longitude
    folium.Marker(
        location=[lat, long],
        draggable=False,
        popup= row["Store Name"],
        tooltip = str(row["Street Address"]) + ", " + str(row["City"]),
        icon=folium.Icon(icon='info-sign', color="green")
        ).add_to(m)


m
```
